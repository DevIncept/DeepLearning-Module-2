# Activation Function

## Binary step function

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.array(range(-10,11))
print(x)
```

```python
# let in binary step function activation value is 3, then any value is grater then 3 is calculated as 1 else 0
def binary_step(x,active_value=3):
    y = []
    for each in x:
        if each > 3:
            y.append(1)
        else:
            y.append(0)
    return y
```

```python
y = binary_step(x,3)
```

```python
def plot_graph(x,y):
    plt.figure(figsize=(10,5))
    plt.plot(x,y)
    plt.xticks(np.arange(-10, 11, step=2))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
```

```python
plot_graph(x,y)
```

## Linear activation function

```python
# let the function is y = mx + c, where m = 0.7 and c = 2
def linear_activation_function(x,m=1,c=0):
    y = (x*m) + c
    return y
```

```python
y = linear_activation_function(x,0.7,2)
```

```python
plot_graph(x,y)
```

## Sigmoid activation function

```python
def sigmoid_function(x):
    return 1/(1 + np.e ** (-x))
```

```python
y = sigmoid_function(x)
```

```python
plot_graph(x,y)
```

## ReLU activation function

```python
def relu_function(x):
    y = []
    for each in x:
        y.append(max(0,each))
    return y
```

```python
y = relu_function(x)
```

```python
plot_graph(x,y)
```
